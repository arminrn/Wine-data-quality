```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()

Scaler.fit(x)
x_scaled = Scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=40)
```

```python
print(f"Shape of X trian: {x_train.shape}");
print(f"Shape of X text: {x_test.shape}");
print(f"Shape of y trian: {y_train.shape}");
print(f"Shape of y text: {y_test.shape}");
```

```python
from sklearn.neighbors import KNeighborsClassifier

modelKNN = KNeighborsClassifier(n_neighbors=6)
```

```python
modelKNN.fit(x_train, y_train);
y_train_predKNN =  modelKNN.predict(x_train)
y_test_predKNN = modelKNN.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score,recall_score
Acc_KNN = accuracy_score(y_test, y_test_predKNN)
F1_Score_KNN = f1_score(y_test, y_test_predKNN, average=None)
Precision_KNN = precision_score(y_test, y_test_predKNN,average=None)
Recall_KNN = recall_score(y_test, y_test_predKNN,average=None)
```

```python
Acc_KNN
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(modelKNN ,x_test,y_test)
```
